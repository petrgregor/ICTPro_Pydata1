Vytvořte kontingenční tabulky z datasetu Chocolate_Sales.csv.

- Spočítejte různé agregační funkce pro sloupce `Amount`, `Boxes Shipped`
- Shlukujte podle `Country`, `Product`

```python
import pandas as pd

df = pd.read_csv('./data/Chocolate_Sales.csv')
df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")
df['Amount'] = df['Amount'].str.replace("$","").str.replace(",","").astype(float)

df.info()
df.head()
```

```python
df_agg = df.groupby(['Country'])['Amount'].agg(['min', 'median', 'mean', 'max', 'var', 'std'])

df_agg
```

```python
df_pivot = df.pivot_table(
    values=['Amount', 'Boxes Shipped'],
    index=['Country', 'Product'],
    #aggfunc=['min', 'median', 'mean', 'max', 'var', 'std']
    aggfunc={
        'Amount': ['min', 'median', 'mean', 'max', 'var', 'std'],
        'Boxes Shipped': ['min', 'median', 'mean', 'max', 'var', 'std']
    }
)

df_pivot
```

## Uložení výsledku

```python
df_pivot.to_csv('./data/Chocolate_Sales_pivot_table.csv')
```
