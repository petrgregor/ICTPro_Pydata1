## Rok, měsíc a den ze sloupce DATE

```python
weather_clean['YEAR'] = weather_clean['DATE'].dt.year
weather_clean['MONTH'] = weather_clean['DATE'].dt.month
weather_clean['DAY'] = weather_clean['DATE'].dt.day
```

## Mapování funkce na sloupec

```python
#def classify_temp(temp):
#    return 'Hot' if temp > 25 else 'Cold'

def classify_temp(temp):
    if temp < -10:
        return 'Extreme freeze'
    if temp < 0:
        return 'Freeze'
    if temp < 20:
        return 'Cold'
    if temp < 30:
        return 'Hot'
    return 'Extreme hot'


weather_clean['TEMP_CLASS'] = weather_clean['TAVG'].apply(classify_temp)
```
