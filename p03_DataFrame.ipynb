# DataFrame

- **DataFrame** je objekt o dvou dimenzích (řádky a sloupce), který můžeme chápat jako tabulku podobnou tabulkám v MS Excel, kde každý sloupec představuje **Series**.
- Položky v **DataFrame** mohou být jakéhokoliv datového typu (číselné, textové, logické, časové atd.), a každý sloupec může mít odlišný datový typ.
- Každá hodnota v **DataFrame** má přidělený **index** (pro řádky) a **sloupec**, což umožňuje přesný přístup k jednotlivým položkám.
- **DataFrame** můžeme chápat jako kolekci více **Series**, kde každý sloupec představuje jednu **Series**.
- U mnoha operací je nutné určit, zda se má operace aplikovat po řádcích nebo po sloupcích. To se definuje pomocí parametru `axis`:
  - `axis=0`: operace se aplikuje přes řádky (po indexu).
  - `axis=1`: operace se aplikuje přes sloupce.

## Vytvoření DataFrame

- **DataFrame** lze vytvořit pomocí konstruktoru `DataFrame()`, kde argumentem je jednodimenzionální nebo dvoudimenzionální datová struktura (např. seznamy, slovníky, numpy pole nebo další pandas objekty). Viz [DataFrame()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html).
- Často používané datové struktury pro inicializaci:
  - Slovník seznamů nebo numpy polí: Klíče představují názvy sloupců a hodnoty seznamů/ polí obsahují data pro daný sloupec.
  - Seznam seznamů (nebo seznam numpy polí): Každý vnitřní seznam představuje jeden řádek.

```python
import pandas as pd

students_dict = {'name': ['Alena', 'Bedřich', 'Ctirad', 'Dušan', 'Eva', 'Filip'],
                 'grade': [1, 2, 1, 3, 3, 2]}
print(f"students_dict: {students_dict}")

students_df = pd.DataFrame(students_dict)
print(f"students_df:\n{students_df}")
```

### Výběr hodnot na základě pozice, štítku či rozsahu hodnot

- Vzhledem k tomu, že **DataFrame** je 2D objekt (obsahuje řádky a sloupce), při výběru hodnot je nutné specifikovat jak **index** (řádek), tak **sloupec**, aby bylo možné přesně určit požadovanou hodnotu.
- K přístupu na hodnoty slouží následující metody:
    - **`loc`**: Slouží pro výběr hodnot na základě **štítků** (label-based). Používá se pro přístup k hodnotám přes řádkové a sloupcové štítky. Logika je stejná jako u **Series**, ale protože se jedná o 2D objekt, je potřeba specifikovat jak řádek, tak sloupec. **Na rozdíl od `iloc` je druhý index (konec rozsahu) zahrnut**, což znamená, že pokud použijete např. `df.loc[0:2]`, vrátí to řádky 0, 1 a 2. Viz [loc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html).
    - **`iloc`**: Slouží pro výběr hodnot na základě **pozice** (integer position-based). Přistupuje se na základě pozic řádků a sloupců, kde se indexuje od nuly. U **`iloc`** je druhý index **vyloučen**. Pokud tedy použijete např. `df.iloc[0:3]`, vrátí to řádky 0, 1 a 2. Viz [iloc](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html).

- Obě metody podporují výběr jednotlivých hodnot, rozsahů (slice), a dokonce i logických podmínek.
    - Například: `df.loc[1, 'column_name']` pro přístup k hodnotě v řádku s indexem `1` a ve sloupci `'column_name'`.
    - Pro rozsahy se používají slices, například: `df.iloc[0:3, 1:4]` pro výběr řádků 0–2 a sloupců 1–3.

- Metody `at` (pro štítky) a `iat` (pro pozice) jsou optimalizované pro rychlý přístup k jednotlivým hodnotám, podobně jako u **Series**, ale i zde je nutné specifikovat jak řádek, tak sloupec.


```python
print(f"students_df:\n{students_df}")

print(f"students_df.iloc[1]:\n{students_df.iloc[1]}")
print(f"students_df.iloc[1:4]:\n{students_df.iloc[1:4]}")
print(f"students_df.iloc[1:4, 1]:\n{students_df.iloc[1:4, 1]}")
print(f"students_df.loc[1:4, 'grade']:\n{students_df.loc[1:4, 'grade']}")
print(f"students_df.loc[1:3, 'name':]:\n{students_df.loc[1:3, 'name':]}")
print(f"students_df.loc[1:3, 'name':'grade]:\n{students_df.loc[1:3, 'name':'grade']}")

print(f"students_df.loc[students_df['grade'] < 3]:\n{students_df.loc[students_df['grade'] < 3]}")
```

## Iterace DataFrame

- **DataFrame** lze iterovat pomocí různých metod, které umožňují procházet jeho řádky nebo sloupce:

    - `iterrows()`: Iteruje přes **řádky** **DataFrame** a vrací dvojice `(index, Series)` pro každý řádek. Každý řádek je reprezentován jako **Series**. Tato metoda je užitečná, pokud potřebujete zpracovat jeden řádek najednou, ale není příliš efektivní pro velké datové sady, protože každý řádek je převeden na **Series**.
      - Použití: `for index, row in df.iterrows():`

    - `items()`: Iteruje přes **sloupce** **DataFrame**. Pro každý sloupec vrací dvojici `(column_name, Series)`, kde `column_name` je název sloupce a `Series` jsou hodnoty ve sloupci. Tato metoda je užitečná, když chcete iterovat přes každý sloupec a pracovat s jeho hodnotami.
      - Použití: `for column_name, column_data in df.items():`

    - `itertuples()`: Iteruje přes **řádky** a pro každý řádek vrací pojmenovaný tuple, což je efektivnější než `iterrows()`. Hodnoty v řádku jsou přístupné jako atributy objektu tuple. Při práci s většími datovými rámci je tato metoda rychlejší než `iterrows()`.
      - Použití: `for row in df.itertuples():`
      - Každý prvek tuple lze přistupovat jako `row.column_name`.

- Iterace přes **DataFrame** je většinou pomalejší než vektorizované operace (které využívají metody pandas pro zpracování dat najednou). Proto je doporučeno, pokud je to možné, upřednostnit vektorové operace nad iterací.



```python
# Iterujeme přes řádky
for i, row in students_df.iterrows():
    if i != 2:
        print(f"Index {i}, row: {row}")

print("-"*80)

# iterujeme přes sloupce
for col, col_data in students_df.items():
    print(f"col = {col}, col_data = {col_data}")

print("-"*80)
# Iterace po řádcích s návratovou hodnotou tuple
for row in students_df.itertuples():
    print(f"row = {row}")
```
