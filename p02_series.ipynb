# Series - úvod

- Objekt o jedné dimenzi (lze si jej představit jako jeden sloupec v tabulce, například v MS Excel).
- Položky v Series mohou být jakéhokoliv datového typu, ale všechny položky v jedné Series obvykle sdílejí stejný datový typ.
- Jednotlivé položky jsou indexovány, tj. každá položka má svůj štítek (index), pomocí kterého na ni lze přistupovat.
- Series může obsahovat chybějící hodnoty (`NaN`, `None`), které mohou ovlivnit analýzu dat nebo výpočty, pokud nejsou správně ošetřeny.

## Vytvoření Series

- Pomocí metody `Series()`, která přijímá jednodimenzionální objekt jako argument (například seznam, numpy pole apod.).
- Výběrem jednoho sloupce z `DataFrame` (výběrem jedné Series z tabulkových dat).
- Je možné explicitně nastavit datový typ jednotlivých položek v Series pomocí parametru `dtype`, což je užitečné pro optimalizaci výkonu a správnost výpočtů.
- Series lze také vytvořit z jiných datových struktur, jako jsou například numpy pole nebo Python slovníky, kde klíče slouží jako index a hodnoty jako položky Series.

```python
import pandas as pd

data = [10, 20, 30, 40, 20]
print(f"data = {data}")

pd_series = pd.Series(data)
print(f"pd_series:\n{pd_series}")

pd_prices = pd.Series(data, name="Price")
print(f"pd_prices:\n{pd_prices}")
```

Informace o Series:

```python
print(f"dtype: {pd_prices.dtype}")  # int64
print(f"shape: {pd_prices.shape}")  # (4,)
print(f"describe(): {pd_prices.describe()}")  # základní charakteristiky
```

## Základní agregační funkce

```python
print(f"{pd_prices.min()}")
price_min = pd_prices.min()
price_max = pd_prices.max()
price_mean = pd_prices.mean()
price_median = pd_prices.median()
price_modus = pd_prices.mode()  # modus vrací Series, protože může nabývat více hodnot
price_count = pd_prices.count()
price_q1 = pd_prices.quantile(0.25)

print(f"\nmin = {price_min}\nmax = {price_max}\nmean = {price_mean}\nmedian = {price_median}\nmodus = {price_modus}\ncount = {price_count}")
print(f"q1 = {price_q1}")
```

## Index Series

- Objekt sloužící jako popisek osy (labels), který identifikuje jednotlivé prvky v datech.
- Nemusí mít unikátní hodnoty, některé metody to však vyžadují, například metody pro spojování nebo indexování s dalšími objekty pandas.
- Jednotlivé hodnoty v indexu jsou neměnné, což znamená, že po vytvoření indexu není možné měnit jednotlivé hodnoty. Celý index lze však nahradit.
- Nastavení vlastního indexu po definici **Series** či **DataFrame** umožňuje metoda `set_axis()`, která slouží k nastavení nové osy (indexu nebo sloupců). Viz [set_axis()](https://pandas.pydata.org/docs/reference/api/pandas.Series.set_axis.html).
- Index lze změnit za pomoci metody `reindex()`, která umožňuje přemapování existujícího indexu na nový nebo přidání chybějících hodnot. Viz [reindex()](https://pandas.pydata.org/docs/reference/api/pandas.Series.reindex.html).
- Ke kontrole, zda jsou hodnoty indexu unikátní, lze použít atribut `is_unique`.

## Datové typy

- Datový typ pro **Series** může být nastaven explicitně při definování nebo odhadnut z dat, která jsou předána jako argument.
- Výchozí datový typ pro **Series** obsahující textové hodnoty nebo hodnoty s různými datovými typy je `object`.
- **Series** obsahující chybějící hodnoty (NaN) bude mít datový typ `float64`, i když se skládá z celých čísel.
- Informace o datovém typu **Series** nebo **DataFrame** je dostupná v atributu `dtype`.
- Přehled datových typů v pandas: [dtypes](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dtypes.html).
- Změna datového typu se provádí pomocí metody `astype()`, viz [astype()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html).
- Pro převod sloupce na datový typ **datetime** slouží metoda `to_datetime()`, viz [to_datetime()](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html).
- Pokud je datovým typem sloupce **datum**, lze přistupovat k atributům data (např. rok, měsíc, den) pomocí atributu `dt`, viz [Series.dt](https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.html).

### Výběr hodnot na základě pozice, štítku či rozsahu hodnot

- Na konkrétní položky **Series** přistupujeme pomocí následujících metod:
    - **`loc`**: Slouží pro výběr hodnot na základě štítků (label-based selection). Používá se, když chceme přistupovat k hodnotám podle konkrétního indexu. **Na rozdíl od `iloc` je druhý index (konec rozsahu) zahrnut**, což znamená, že pokud použijeme např. `series.loc[0:2]`, vrátí to hodnoty s indexy 0, 1 a 2. Viz [loc](https://pandas.pydata.org/docs/reference/api/pandas.Series.loc.html).
    - **`iloc`**: Slouží pro výběr hodnot na základě pozice (integer position-based selection). Používá se při přístupu na základě pozic, kde se indexuje od nuly. **Druhý index je vyloučen**, takže pokud použijeme např. `series.iloc[0:3]`, vrátí to hodnoty s indexy 0, 1 a 2. Viz [iloc](https://pandas.pydata.org/docs/reference/api/pandas.Series.iloc.html).
    - **`at`**: Preferovaná metoda při přístupu k jediné hodnotě na základě štítku (label), rychlejší než `loc` při práci s jednou hodnotou. Viz [at](https://pandas.pydata.org/docs/reference/api/pandas.Series.at.html).
    - **`iat`**: Slouží k přístupu na jedinou hodnotu pomocí pozice (indexu). Je analogií k `at`, ale používá se při pozicovém přístupu. Viz [iat](https://pandas.pydata.org/docs/reference/api/pandas.Series.iat.html).

- Přistupovat lze na konkrétní položky či rozsahy položek, přičemž na rozsah položek přistupujeme skrz tzv. **slice**. Slice je výběr podmnožiny hodnot, kde začátek a konec výběru jsou specifikovány. Viz [slicing ranges](https://pandas.pydata.org/docs/user_guide/indexing.html#slicing-ranges).
- Pro výběr na základě složitějších podmínek (např. logické podmínky, masky) lze použít tzv. **boolean indexing**, který umožňuje filtrovat řádky na základě podmínek.

```python
print(f"{pd_prices.iloc[2]}")
print(f"{pd_series.iloc[1:4]}")
print(f"{pd_series.iloc[2:]}")
print(f"{pd_series.iloc[:3]}")
print(f"{pd_series.iloc[::-1]}")
```

#### Explicitní je lepší než implicitní

- Indexování bez uvedení operátorů loc či iloc má různé chování v případě, kdy je na indexu číslo a text, viz níže.
- Proto je žádoucí používat operátory loc, iloc či at

```python
import numpy as np
s1 = pd.Series(np.arange(3), index=[100, 101, 102])
print(f"s1:\n{s1}")

s2 = pd.Series(np.arange(3), index=['a', 'b', 'c'])
print(f"s2:\n{s2}")

#print(s1[1])  # KeyError
print(s1[101])

#print(s2[1])  # KeyError
print(s2['b'])

#loc
#print(s1.loc[1])  # KeyError

#iloc
print(s1.iloc[1])
```

## Filtrování Series
- Pro filtrování **Series** lze použít klasické operátory porovnání (`==`, `!=`, `<`, `>`, `<=`, `>=`), viz [Python value comparisons](https://docs.python.org/3/reference/expressions.html#value-comparisons).
- Pořadí operátorů při řetězení podmínek (priorita operací) je klíčové a řídí se pravidly definovanými v Pythonu. Viz [Python operator precedence](https://docs.python.org/3/reference/expressions.html#operator-precedence).
- Na rozdíl od klasického Pythonu, kde se používají `and` a `or` pro logické podmínky, v pandas se pro filtrování používají následující logické operátory:
    - `&` (logická operace "a" místo `and`)
    - `|` (logická operace "nebo" místo `or`)
    - Je potřeba jednotlivé podmínky obalit závorkami, například `(Series > 5) & (Series < 10)`.
- K filtrování textových hodnot či hodnot s přibližnou shodou (např. při hledání substringů) lze použít metodu `filter()`. Viz [filter()](https://pandas.pydata.org/docs/reference/api/pandas.Series.filter.html).
- Data lze také filtrovat pomocí booleanovského vektoru, který umožňuje vracet pouze hodnoty splňující zadané podmínky. Viz [boolean indexing](https://pandas.pydata.org/docs/user_guide/indexing.html#boolean-indexing).

### Booleanovský vektor
- **Booleanovský vektor** je série logických hodnot (`True` nebo `False`), kde každá hodnota označuje, zda odpovídající položka splňuje podmínku.
- Při filtrování se vrací pouze položky, kde podmínka vrací `True`.
- Hodnoty `NA` (chybějící hodnoty) se považují za `False`, takže nejsou zahrnuty do výsledného filtru.

```python
print(f"Prices:\n{pd_prices}")

pd_prices_lt_30 = pd_prices < 30
print(f"pd_prices < 30:\n{pd_prices_lt_30}")

pd_prices_lt_30_values = pd_prices[pd_prices_lt_30]
print(f"Prices < 30:\n{pd_prices_lt_30_values}")
pd_prices_lt_30_values = pd_prices[pd_prices < 30]
print(f"Prices < 30:\n{pd_prices_lt_30_values}")

print(f"Prices >10 AND < 40:\n{pd_prices[(pd_prices > 10) & (pd_prices < 40)]}")
print(f"Prices < 20 OR > 30:\n{pd_series[(pd_series < 20) | (pd_series > 30)]}")
```
