```python
import pandas as pd
from pandas.conftest import dropna

weather = pd.read_csv('./data/pydata_data.csv', sep=";")

weather['YEAR'] = weather['DATE'].str[:4].astype(int)
weather['MONTH'] = weather['DATE'].str[5:7].astype(int)
weather['DAY'] = weather['DATE'].str[8:].astype(int)

weather['DATE'] = pd.to_datetime(weather['DATE'])

weather.info()
weather.head()
```

# Statistiky

```python
weather.describe()
```

```python
print(f"Šikmost: {weather['TAVG'].skew()}")
print(f"Špičatost: {weather['TAVG'].kurt()}")
```

## Ověření normality

```python
from scipy import stats

stat, p_value = stats.shapiro(weather['TMIN'], nan_policy='omit')

print(f"stat = {stat}, p_value = {p_value}")
```
