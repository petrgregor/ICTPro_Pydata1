## Načtení dat ✅

```python
import pandas as pd

df = pd.read_csv('./data/Chocolate_Sales.csv')

df.info()
df.head()
```

## Odstranění nepotřebných sloupců ✅

Zde asi není potřeba nic odstraňovat.

## Transformace na vhodné datové typy ✅

```python
df.info()
df.head()

df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")

df.info()
df.head()
```

```python
#def clean_amount(amount: str) -> str:
#    return amount.replace("$","").replace(",","")

df['Amount'] = df['Amount'].str.replace("$","").str.replace(",","").astype(float)

df.info()
df.head()
```

```python
df['Country'] = df['Country'].astype('category')

df.info()
df.head()
```

```python
df['Product'] = df['Product'].astype('category')

df.info()
df.head()
```

## Řešení chybějících hodnot ✅

```python
print(df.isnull().sum())
```

Žádné chybějící hodnoty.

## Odstranění duplicitních řádků ✅

```python
print(f"Počet duplicitních řádků: {df.duplicated().sum()}")
```

Žádné duplicitní řídky.

## Popisná statistika

```python
df.describe()
```

## Uložení upravených dat

```python
df.to_csv('./data/Chocolate_Sales_cleaned.csv', index=False)
```
