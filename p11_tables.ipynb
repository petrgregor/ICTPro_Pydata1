```python
import pandas as pd

weather = pd.read_csv('./data/pydata_data.csv', sep=";")

weather['YEAR'] = weather['DATE'].str[:4].astype(int)
weather['MONTH'] = weather['DATE'].str[5:7].astype(int)
weather['DAY'] = weather['DATE'].str[8:].astype(int)

weather['DATE'] = pd.to_datetime(weather['DATE'])

weather.info()
weather.head()
```

```python
corr_matrix = weather.corr(numeric_only = True)
print(f"Pearson:\n{corr_matrix}")

corr_matrix.info()
#corr_matrix_styled = corr_matrix.style.highlight_max(color="red")
corr_matrix_styled = (corr_matrix.style.highlight_max(color="white")
                      .highlight_between(left=0.7, right=0.9999, color="red")
                      .highlight_between(left=0.3, right=0.7, color="orange")
                      .highlight_between(left=-1, right=-0.7, color="blue")
                      .highlight_between(left=-0.7, right=-0.3, color="#3262a8"))
display(corr_matrix_styled)
```

```python
corr_matrix_styled = corr_matrix.style.bar(color="red")
display(corr_matrix_styled)
```

```python
corr_matrix_styled = corr_matrix.style.format('{:.3}')
display(corr_matrix_styled)
```
