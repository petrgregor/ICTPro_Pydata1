# Agregace

- **Základní agregace** se provádí pomocí metod `agg()` a `aggregate()`, které umožňují aplikovat jednu nebo více agregačních funkcí současně (např. `sum()`, `mean()`, `min()`, `max()`). Tyto metody jsou flexibilní a podporují kombinování vestavěných a vlastních funkcí. Viz [agg()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html) a [aggregate()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.aggregate.html).

- **Kumulované agregace** jsou operace, které aplikují agregační funkci postupně na řadu hodnot. Mezi nejčastěji používané kumulované metody patří:
    - `cummax()`: Kumulativní maximum
    - `cummin()`: Kumulativní minimum
    - `cumsum()`: Kumulativní součet


- **Klouzavé agregace** se provádějí pomocí metody `rolling()`, která vytváří klouzavé okno o specifikované velikosti a aplikuje agregační funkce na jednotlivá okna. To je užitečné pro analýzu trendů a výpočty založené na časových řadách. Viz [rolling()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html).

### Shlukování dat za pomoci metody `groupby()`

- Metoda `groupby()` slouží ke shlukování dat podle jednoho nebo více klíčů. Lze ji použít k výpočtu souhrnných statistik pro jednotlivé skupiny. V případě **DataFrame** můžeme shlukovat buď podle řádků (standardně `axis=0`), nebo podle sloupců (`axis=1`). Shlukování lze provádět na základě hodnot ve sloupci, na indexech nebo i na základě více klíčů. Viz [groupby()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html).

### Transformace dat

- **Transformace řádků na sloupce a opačně** se provádí pomocí metody `transpose()`, která přehazuje řádky za sloupce a naopak. Tato operace se často používá k reorganizaci dat v **DataFrame**. Viz [transpose()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transpose.html).

- **Transformace řádků a sloupců s unikátními hodnotami** pomocí metody `pivot()`. Tato metoda přehazuje data na základě konkrétních hodnot sloupce a vrací **DataFrame** s novými indexy a sloupci, ale nepodporuje agregaci. Je užitečná pro transformaci dat bez nutnosti jejich souhrnného zpracování. Viz [pivot()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html).
    - Pozn.: `pivot()` nepodporuje agregace; pokud se v datech nacházejí duplicitní hodnoty, vrací chybu.

- **Transformace řádků a sloupců s agregací** pomocí metody `pivot_table()`. Tato metoda podporuje agregaci hodnot pomocí specifikovaných funkcí (např. průměr, součet) a je vhodná pro vytváření kontingenčních tabulek. Viz [pivot_table()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot_table.html).
    - Příklad: `df.pivot_table(values='hodnota', index='index_sloupec', columns='sloupec', aggfunc='sum')`


```python
import pandas as pd

weather = pd.read_csv('./data/pydata_data.csv', sep=";")

weather['YEAR'] = weather['DATE'].str[:4].astype(int)
weather['MONTH'] = weather['DATE'].str[5:7].astype(int)
weather['DAY'] = weather['DATE'].str[8:].astype(int)

weather['DATE'] = pd.to_datetime(weather['DATE'])
#weather['YEAR'] = weather['DATE'].dt.year
#weather['MONTH'] = weather['DATE'].dt.month
#weather['DAY'] = weather['DATE'].dt.day

weather.info()
weather.head()
```

## Agregační funkce
### Jedna agregační funkce

```python
temp_avg = weather.groupby('NAME')['TAVG'].mean()
temp_avg
```

### Více agregačních funkcí

```python
temp_by_name = weather.groupby('NAME')['TAVG'].agg(['min', 'mean', 'median', 'max'])
temp_by_name
```

### Shlukování podle více kategorií

```python
temp_by_name_year = weather.groupby(['NAME', 'YEAR'])['TAVG'].agg(['min', 'mean', 'median', 'max'])
temp_by_name_year
```

## Pokročilá agregace
Různé agregační funkce pro různé sloupce.

```python
weather_agg = weather.groupby(['NAME', 'YEAR']).agg({'TMIN': 'min',
                                                     'TMAX': 'max',
                                                     'TAVG': 'mean',
                                                     'PRCP': 'sum',
                                                     'STATION': 'nunique'})

weather_agg
```

### pivot_table()

```python
weather_prcp = weather.pivot_table(values='PRCP',
                                   index='NAME',
                                   columns='YEAR',
                                   aggfunc='mean',
                                   dropna=True,
                                   fill_value='NA')

weather_prcp
```

```python
weather_prcp = weather.pivot_table(values=['PRCP', 'TAVG'],
                                   index=['NAME', 'YEAR'],
                                   #columns='YEAR',
                                   aggfunc=['mean', 'min', 'max'],
                                   dropna=True,
                                   fill_value='NA')

weather_prcp
```
